```python
import cv2
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

from pathlib import Path
from insightface.app.face_analysis import FaceAnalysis
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import euclidean_distances

# =========================
# 1. PATHS
# =========================
FOLDERS = {
    "uncanny": Path(r"C:\Users\maryc\Downloads\881project\uncannyrobot"),
    "stylized": Path(r"C:\Users\maryc\Downloads\881project\stylishedrobot"),
    "human_conventional": Path(r"C:\Users\maryc\Downloads\881project\conventionalfaces"),
    "human_unconventional": Path(r"C:\Users\maryc\Downloads\881project\unconventionalfaces"),
}

# save cropped faces here
CROPPED_ROOT = Path(r"C:\Users\maryc\Downloads\881project\cropped_faces")
CROPPED_ROOT.mkdir(exist_ok=True)

# how many to use
MAX_PER_GROUP = {
    "uncanny": None,                # use all
    "stylized": None,               # use all
    "human_conventional": 100,      # sample 100
    "human_unconventional": None,   # use all
}

IMG_EXTS = {".jpg", ".jpeg", ".png", ".webp"}

# =========================
# 2. HELPERS
# =========================
def list_images(folder, max_images=None, seed=42):
    files = [p for p in folder.iterdir() if p.suffix.lower() in IMG_EXTS]
    files = sorted(files)
    if max_images is not None and len(files) > max_images:
        random.seed(seed)
        files = random.sample(files, max_images)
    return files

def crop_with_margin(img, bbox, margin_ratio=0.2):
    h, w = img.shape[:2]
    x1, y1, x2, y2 = map(int, bbox)

    bw = x2 - x1
    bh = y2 - y1

    mx = int(bw * margin_ratio)
    my = int(bh * margin_ratio)

    x1 = max(0, x1 - mx)
    y1 = max(0, y1 - my)
    x2 = min(w, x2 + mx)
    y2 = min(h, y2 + my)

    return img[y1:y2, x1:x2]

def save_crop(out_path, crop, size=(224, 224)):
    crop = cv2.resize(crop, size, interpolation=cv2.INTER_CUBIC)
    cv2.imwrite(str(out_path), crop)

# =========================
# 3. LOAD MODEL
# =========================
app = FaceAnalysis(name="buffalo_l")
app.prepare(ctx_id=-1)  # CPU

def get_embedding_and_crop(path, app, save_dir=None):
    img = cv2.imread(str(path))
    if img is None:
        return None, "read_fail"

    faces = app.get(img)
    if len(faces) == 0:
        return None, "no_face"

    faces = sorted(
        faces,
        key=lambda f: (f.bbox[2] - f.bbox[0]) * (f.bbox[3] - f.bbox[1]),
        reverse=True
    )
    face = faces[0]

    crop = crop_with_margin(img, face.bbox, margin_ratio=0.2)

    if save_dir is not None:
        save_dir.mkdir(parents=True, exist_ok=True)
        save_crop(save_dir / path.name, crop, size=(224, 224))

    return face.embedding.astype(np.float32), "ok"

# =========================
# 4. EXTRACT EMBEDDINGS
# =========================
rows = []
embs = []

for group, folder in FOLDERS.items():
    paths = list_images(folder, max_images=MAX_PER_GROUP[group])
    print(f"{group}: selected {len(paths)} images")

    group_crop_dir = CROPPED_ROOT / group

    for p in paths:
        emb, status = get_embedding_and_crop(p, app, save_dir=group_crop_dir)
        rows.append({
            "group": group,
            "path": str(p),
            "status": status
        })
        if status == "ok":
            embs.append(emb)

df = pd.DataFrame(rows)

print("\nStatus counts:")
print(df["status"].value_counts())

print("\nGroup x Status:")
print(df.groupby(["group", "status"]).size())

# save logs
df.to_csv("rq1_embedding_status_all.csv", index=False)
df[df["status"] != "ok"].to_csv("rq1_embedding_failures.csv", index=False)

# keep only successful samples
df_ok = df[df["status"] == "ok"].copy().reset_index(drop=True)
X = np.vstack(embs)

df_ok.to_csv("rq1_embeddings_meta.csv", index=False)
np.save("rq1_embeddings.npy", X)

print(f"\nSuccessful embeddings: {len(df_ok)}")

# =========================
# 5. PCA VISUALIZATION
# =========================
groups = df_ok["group"].tolist()

pca = PCA(n_components=2, random_state=0)
X2 = pca.fit_transform(X)

colors = {
    "human_conventional": "blue",
    "human_unconventional": "purple",
    "uncanny": "orange",
    "stylized": "green"
}

plt.figure(figsize=(8, 6))
for g in ["human_conventional", "human_unconventional", "uncanny", "stylized"]:
    idx = [i for i, gg in enumerate(groups) if gg == g]
    if idx:
        plt.scatter(
            X2[idx, 0],
            X2[idx, 1],
            label=g,
            s=30,
            alpha=0.8,
            c=colors[g]
        )

plt.legend()
plt.title("RQ1 PCA of Face Embeddings")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.savefig("rq1_pca_unbalanced.png", dpi=200)
plt.show()

print("Saved: rq1_pca_unbalanced.png")

# =========================
# 6. SPLIT EMBEDDINGS
# =========================
mask_unc = (df_ok["group"] == "uncanny").to_numpy()
mask_sty = (df_ok["group"] == "stylized").to_numpy()
mask_hc  = (df_ok["group"] == "human_conventional").to_numpy()
mask_hu  = (df_ok["group"] == "human_unconventional").to_numpy()

unc_X = X[mask_unc]
sty_X = X[mask_sty]
hc_X  = X[mask_hc]
hu_X  = X[mask_hu]

unc_paths = df_ok.loc[mask_unc, "path"].tolist()
hc_paths  = df_ok.loc[mask_hc, "path"].tolist()
hu_paths  = df_ok.loc[mask_hu, "path"].tolist()

print("\nShapes:")
print("uncanny:", unc_X.shape)
print("stylized:", sty_X.shape)
print("human_conventional:", hc_X.shape)
print("human_unconventional:", hu_X.shape)

# =========================
# 7. CENTROID DISTANCE TEST
# =========================
hc_centroid = hc_X.mean(axis=0, keepdims=True)
hu_centroid = hu_X.mean(axis=0, keepdims=True)

dist_unc_to_hc = euclidean_distances(unc_X, hc_centroid).reshape(-1)
dist_unc_to_hu = euclidean_distances(unc_X, hu_centroid).reshape(-1)

centroid_df = pd.DataFrame({
    "uncanny_path": unc_paths,
    "dist_to_conventional_centroid": dist_unc_to_hc,
    "dist_to_unconventional_centroid": dist_unc_to_hu
})

centroid_df["closer_to"] = np.where(
    centroid_df["dist_to_unconventional_centroid"] < centroid_df["dist_to_conventional_centroid"],
    "human_unconventional",
    "human_conventional"
)

centroid_df.to_csv("rq1_uncanny_centroid_distances.csv", index=False)

print("\nMean centroid distances for uncanny robots:")
print("to conventional humans:", dist_unc_to_hc.mean())
print("to unconventional humans:", dist_unc_to_hu.mean())

print("\nCentroid comparison:")
print(centroid_df["closer_to"].value_counts())

# =========================
# 8. NEAREST-NEIGHBOR TEST
# =========================
D_to_hc = euclidean_distances(unc_X, hc_X)
D_to_hu = euclidean_distances(unc_X, hu_X)

nearest_hc_dist = D_to_hc.min(axis=1)
nearest_hu_dist = D_to_hu.min(axis=1)

nearest_hc_idx = D_to_hc.argmin(axis=1)
nearest_hu_idx = D_to_hu.argmin(axis=1)

nn_df = pd.DataFrame({
    "uncanny_path": unc_paths,
    "nearest_conventional_path": [hc_paths[i] for i in nearest_hc_idx],
    "nearest_conventional_dist": nearest_hc_dist,
    "nearest_unconventional_path": [hu_paths[i] for i in nearest_hu_idx],
    "nearest_unconventional_dist": nearest_hu_dist
})

nn_df["nearest_group"] = np.where(
    nn_df["nearest_unconventional_dist"] < nn_df["nearest_conventional_dist"],
    "human_unconventional",
    "human_conventional"
)

nn_df.to_csv("rq1_uncanny_nearest_neighbor.csv", index=False)

print("\nNearest-neighbor comparison:")
print(nn_df["nearest_group"].value_counts())

# =========================
# 9. SIMPLE RQ1 SUMMARY
# =========================
n_unc = len(nn_df)

n_nn_unconv = (nn_df["nearest_group"] == "human_unconventional").sum()
n_nn_conv   = (nn_df["nearest_group"] == "human_conventional").sum()

n_cent_unconv = (centroid_df["closer_to"] == "human_unconventional").sum()
n_cent_conv   = (centroid_df["closer_to"] == "human_conventional").sum()

print("\n================ RQ1 SUMMARY ================")
print(f"Total uncanny robots analyzed: {n_unc}")
print(f"Nearest-neighbor closer to unconventional humans: {n_nn_unconv}")
print(f"Nearest-neighbor closer to conventional humans:   {n_nn_conv}")
print(f"Centroid closer to unconventional humans:        {n_cent_unconv}")
print(f"Centroid closer to conventional humans:          {n_cent_conv}")

if n_nn_unconv > n_nn_conv:
    print("\nPreliminary answer to RQ1:")
    print("Uncanny robot faces appear closer to unconventional human faces than to conventional human faces.")
else:
    print("\nPreliminary answer to RQ1:")
    print("Uncanny robot faces do not appear closer to unconventional human faces than to conventional human faces.")

# =========================
# 10. OPTIONAL SUMMARY BAR CHART
# =========================
summary_counts = pd.DataFrame({
    "Comparison": ["Centroid", "Centroid", "Nearest Neighbor", "Nearest Neighbor"],
    "Closer To": ["Conventional", "Unconventional", "Conventional", "Unconventional"],
    "Count": [n_cent_conv, n_cent_unconv, n_nn_conv, n_nn_unconv]
})

plt.figure(figsize=(7, 5))
positions = [0, 1, 3, 4]
plt.bar(positions, summary_counts["Count"])
plt.xticks(positions, ["Conv.", "Unconv.", "Conv.", "Unconv."])
plt.ylabel("Count of uncanny robots")
plt.title("RQ1: Which human group are uncanny robots closer to?")
plt.tight_layout()
plt.savefig("rq1_summary_bar_unbalanced.png", dpi=200)
plt.show()

print("Saved: rq1_summary_bar_unbalanced.png")
```

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from insightface.app.face_analysis import FaceAnalysis
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ======================================
# 1. CHECK BALANCED DATA EXISTS
# ======================================
print(df_bal["group"].value_counts())
print("Balanced embedding shape:", X_bal.shape)

# ======================================
# 2. LOAD FACE MODEL
# ======================================
app = FaceAnalysis(name="buffalo_l")
app.prepare(ctx_id=-1)  # CPU

# ======================================
# 3. HELPERS
# ======================================
def crop_with_margin(img, bbox, margin_ratio=0.2):
    h, w = img.shape[:2]
    x1, y1, x2, y2 = map(int, bbox)

    bw = x2 - x1
    bh = y2 - y1

    mx = int(bw * margin_ratio)
    my = int(bh * margin_ratio)

    x1 = max(0, x1 - mx)
    y1 = max(0, y1 - my)
    x2 = min(w, x2 + mx)
    y2 = min(h, y2 + my)

    return img[y1:y2, x1:x2]

def get_face_crop(path):
    img = cv2.imread(str(path))
    if img is None:
        return None, "read_fail"

    faces = app.get(img)
    if len(faces) == 0:
        return None, "no_face"

    faces = sorted(
        faces,
        key=lambda f: (f.bbox[2] - f.bbox[0]) * (f.bbox[3] - f.bbox[1]),
        reverse=True
    )
    face = faces[0]
    crop = crop_with_margin(img, face.bbox, margin_ratio=0.2)
    crop = cv2.resize(crop, (224, 224), interpolation=cv2.INTER_CUBIC)
    gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
    return gray, "ok"

def symmetry_score(img):
    h, w = img.shape
    mid = w // 2

    left = img[:, :mid]
    right = img[:, w-mid:]
    right_flipped = cv2.flip(right, 1)

    min_w = min(left.shape[1], right_flipped.shape[1])
    left = left[:, :min_w]
    right_flipped = right_flipped[:, :min_w]

    diff = np.abs(left.astype(np.float32) - right_flipped.astype(np.float32))
    return diff.mean()  # lower = more symmetric

def texture_variance(img):
    lap = cv2.Laplacian(img, cv2.CV_64F)
    return lap.var()

def edge_density(img):
    edges = cv2.Canny(img, 100, 200)
    return edges.mean()

def intensity_variance(img):
    return img.var()

def center_border_difference(img):
    h, w = img.shape
    ch1, ch2 = h // 4, 3 * h // 4
    cw1, cw2 = w // 4, 3 * w // 4

    center = img[ch1:ch2, cw1:cw2]

    border_mask = np.ones_like(img, dtype=bool)
    border_mask[ch1:ch2, cw1:cw2] = False
    border = img[border_mask]

    return abs(center.mean() - border.mean())

# ======================================
# 4. EXTRACT RQ2 FEATURES FROM BALANCED SET
# ======================================
rq2_rows = []

for _, row in df_bal.iterrows():
    path = row["path"]
    group = row["group"]

    crop, status = get_face_crop(path)
    if status != "ok":
        rq2_rows.append({
            "group": group,
            "path": path,
            "status": status
        })
        continue

    rq2_rows.append({
        "group": group,
        "path": path,
        "status": "ok",
        "symmetry_diff": symmetry_score(crop),
        "texture_var": texture_variance(crop),
        "edge_density": edge_density(crop),
        "intensity_var": intensity_variance(crop),
        "center_border_diff": center_border_difference(crop),
    })

rq2_df = pd.DataFrame(rq2_rows)

print("\nRQ2 status counts:")
print(rq2_df["status"].value_counts())

print("\nRQ2 group x status:")
print(rq2_df.groupby(["group", "status"]).size())

rq2_df.to_csv("rq2_balanced_feature_status.csv", index=False)

# keep only successful crops
rq2_ok = rq2_df[rq2_df["status"] == "ok"].copy().reset_index(drop=True)

print("\nRQ2 valid per group:")
print(rq2_ok["group"].value_counts())

# ======================================
# 5. BUILD CONSISTENCY SCORE
# ======================================
feature_cols = [
    "symmetry_diff",
    "texture_var",
    "edge_density",
    "intensity_var",
    "center_border_diff"
]

X_feat = rq2_ok[feature_cols].to_numpy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_feat)

# PCA on consistency features
pca_feat = PCA(n_components=2, random_state=0)
X_feat_pca = pca_feat.fit_transform(X_scaled)

rq2_ok["pc1"] = X_feat_pca[:, 0]
rq2_ok["pc2"] = X_feat_pca[:, 1]

# Consistency score:
# distance from global feature center
# smaller distance = more internally typical / consistent
center = X_scaled.mean(axis=0, keepdims=True)
dist_from_center = np.linalg.norm(X_scaled - center, axis=1)

rq2_ok["consistency_score"] = -dist_from_center  # higher = more consistent

rq2_ok.to_csv("rq2_balanced_features_with_score.csv", index=False)

# ======================================
# 6. PCA PLOT OF CONSISTENCY FEATURES
# ======================================
colors = {
    "human_conventional": "blue",
    "human_unconventional": "purple",
    "uncanny": "orange",
    "stylized": "green"
}

plt.figure(figsize=(8, 6))
for g in ["human_conventional", "human_unconventional", "uncanny", "stylized"]:
    sub = rq2_ok[rq2_ok["group"] == g]
    if len(sub) > 0:
        plt.scatter(
            sub["pc1"],
            sub["pc2"],
            label=g,
            s=35,
            alpha=0.8,
            c=colors[g]
        )

plt.legend()
plt.title("RQ2 PCA of Consistency Features (balanced set)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.savefig("rq2_balanced_pca_consistency.png", dpi=200)
plt.show()

# ======================================
# 7. BAR PLOT OF MEAN CONSISTENCY
# ======================================
group_means = rq2_ok.groupby("group")["consistency_score"].mean().sort_values()

plt.figure(figsize=(7, 5))
plt.bar(group_means.index, group_means.values)
plt.ylabel("Mean consistency score")
plt.title("RQ2: Consistency by group (balanced set)")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig("rq2_balanced_consistency_bar.png", dpi=200)
plt.show()

print("\nMean consistency score by group:")
print(group_means)

# ======================================
# 8. SIMPLE RQ2 SUMMARY
# ======================================
unc_mean = rq2_ok[rq2_ok["group"] == "uncanny"]["consistency_score"].mean()
sty_mean = rq2_ok[rq2_ok["group"] == "stylized"]["consistency_score"].mean()
hc_mean  = rq2_ok[rq2_ok["group"] == "human_conventional"]["consistency_score"].mean()
hu_mean  = rq2_ok[rq2_ok["group"] == "human_unconventional"]["consistency_score"].mean()

print("\n================ RQ2 SUMMARY ================")
print("uncanny mean consistency:", unc_mean)
print("stylized mean consistency:", sty_mean)
print("human_conventional mean consistency:", hc_mean)
print("human_unconventional mean consistency:", hu_mean)

if unc_mean < sty_mean and unc_mean < hc_mean and unc_mean < hu_mean:
    print("\nPreliminary answer to RQ2:")
    print("Uncanny robot faces show the lowest internal consistency among the balanced groups.")
else:
    print("\nPreliminary answer to RQ2:")
    print("Uncanny robot faces do not show the lowest consistency among all balanced groups.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import euclidean_distances

# ======================================
# 1. CHECK INPUTS
# ======================================
print("Balanced counts:")
print(df_bal["group"].value_counts())

print("\nRQ2 counts:")
print(rq2_ok["group"].value_counts())

# ======================================
# 2. MERGE CONSISTENCY SCORES INTO BALANCED DATA
# ======================================
rq2_scores = rq2_ok[["group", "path", "consistency_score"]].copy()

rq3_df = df_bal.merge(rq2_scores, on=["group", "path"], how="left")

print("\nMissing consistency scores:", rq3_df["consistency_score"].isna().sum())

rq3_df = rq3_df.dropna(subset=["consistency_score"]).copy().reset_index(drop=True)

print("\nRQ3 usable counts:")
print(rq3_df["group"].value_counts())

# ======================================
# 3. ALIGN EMBEDDINGS WITHOUT orig_idx
# ======================================
# Build a lookup from (group, path) -> embedding row in df_bal/X_bal
key_to_idx = {
    (row["group"], row["path"]): i
    for i, row in df_bal.reset_index(drop=True).iterrows()
}

rq3_indices = []
missing_keys = []

for _, row in rq3_df.iterrows():
    key = (row["group"], row["path"])
    if key in key_to_idx:
        rq3_indices.append(key_to_idx[key])
    else:
        missing_keys.append(key)

if missing_keys:
    print("\nWarning: missing keys during alignment:", len(missing_keys))

X_rq3 = np.vstack([X_bal[i] for i in rq3_indices])

# ======================================
# 4. DISTANCE TO CONVENTIONAL-HUMAN CENTROID
# ======================================
mask_hc = (rq3_df["group"] == "human_conventional").to_numpy()
mask_hu = (rq3_df["group"] == "human_unconventional").to_numpy()

hc_X = X_rq3[mask_hc]
hu_X = X_rq3[mask_hu]

hc_centroid = hc_X.mean(axis=0, keepdims=True)
hu_centroid = hu_X.mean(axis=0, keepdims=True)

rq3_df["dist_to_conventional_centroid"] = euclidean_distances(X_rq3, hc_centroid).reshape(-1)
rq3_df["dist_to_unconventional_centroid"] = euclidean_distances(X_rq3, hu_centroid).reshape(-1)

rq3_df.to_csv("rq3_group_metrics.csv", index=False)

# ======================================
# 5. GROUP MEANS
# ======================================
rq3_group_means = rq3_df.groupby("group")[[
    "dist_to_conventional_centroid",
    "dist_to_unconventional_centroid",
    "consistency_score"
]].mean()

print("\nRQ3 group means:")
print(rq3_group_means)

# ======================================
# 6. BAR CHARTS
# ======================================
plt.figure(figsize=(7,5))
dist_conv_means = rq3_group_means["dist_to_conventional_centroid"].sort_values()
plt.bar(dist_conv_means.index, dist_conv_means.values)
plt.ylabel("Mean distance to conventional-human centroid")
plt.title("RQ3: Conventional realism constraint proxy")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig("rq3_distance_to_conventional_bar.png", dpi=200)
plt.show()

plt.figure(figsize=(7,5))
cons_means = rq3_group_means["consistency_score"].sort_values()
plt.bar(cons_means.index, cons_means.values)
plt.ylabel("Mean consistency score")
plt.title("RQ3: Internal consistency by group")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig("rq3_consistency_bar.png", dpi=200)
plt.show()

# ======================================
# 7. TWO-METRIC SCATTER
# ======================================
colors = {
    "human_conventional": "blue",
    "human_unconventional": "purple",
    "uncanny": "orange",
    "stylized": "green"
}

plt.figure(figsize=(8,6))

for g in ["human_conventional", "human_unconventional", "uncanny", "stylized"]:
    row = rq3_group_means.loc[g]
    plt.scatter(
        row["dist_to_conventional_centroid"],
        row["consistency_score"],
        s=180,
        c=colors[g],
        label=g
    )
    plt.text(
        row["dist_to_conventional_centroid"] + 0.01,
        row["consistency_score"] + 0.01,
        g,
        fontsize=9
    )

plt.xlabel("Distance to conventional-human centroid\n(higher = lower conventional realism constraint)")
plt.ylabel("Consistency score\n(higher = more internally coherent)")
plt.title("RQ3: Constraint vs consistency")
plt.tight_layout()
plt.savefig("rq3_constraint_vs_consistency_scatter.png", dpi=200)
plt.show()

# ======================================
# 8. SIMPLE RQ3 SUMMARY
# ======================================
unc_dist = rq3_group_means.loc["uncanny", "dist_to_conventional_centroid"]
unc_cons = rq3_group_means.loc["uncanny", "consistency_score"]

hu_dist = rq3_group_means.loc["human_unconventional", "dist_to_conventional_centroid"]
hu_cons = rq3_group_means.loc["human_unconventional", "consistency_score"]

sty_dist = rq3_group_means.loc["stylized", "dist_to_conventional_centroid"]
sty_cons = rq3_group_means.loc["stylized", "consistency_score"]

print("\n================ RQ3 SUMMARY ================")
print("Uncanny robots:")
print("  distance to conventional centroid:", unc_dist)
print("  consistency:", unc_cons)

print("Unconventional humans:")
print("  distance to conventional centroid:", hu_dist)
print("  consistency:", hu_cons)

print("Stylized robots:")
print("  distance to conventional centroid:", sty_dist)
print("  consistency:", sty_cons)

if hu_cons > unc_cons and hu_dist >= unc_dist:
    print("\nPreliminary answer to RQ3:")
    print("Unconventional-human targets appear to reduce realism constraints relative to conventional realism attempts, because they maintain higher consistency while being at least as far from the conventional-human target.")
elif hu_cons > unc_cons:
    print("\nPreliminary answer to RQ3:")
    print("Unconventional-human targets show higher consistency than conventional realism attempts, which supports reduced realism burden, although their distance from the conventional-human target is not greater.")
else:
    print("\nPreliminary answer to RQ3:")
    print("The current balanced analysis does not clearly show that unconventional-human targets reduce realism constraints relative to conventional realism attempts.")
```
